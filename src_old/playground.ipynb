```python
# open image files
import numpy as np
import glob
from PIL import Image
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern
from sklearn.cluster import KMeans
from scipy import ndimage

d = 'sargentii_17'

files = []
for file in glob.glob(d+'/*.ti*'):
    files.append(file)

# open first image to get the image dimensions
im = np.array(Image.open(files[0]))

# define image_stack array
image_stack = np.zeros([len(files),np.shape(im)[0],np.shape(im)[1]])
# read and standardize all images
for i in range(len(files)):
    im = np.array(Image.open(files[i]))        

    image_stack[i,:,:] = im
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
matplotlib.rcParams.update({'font.size': 16})
matplotlib.rcParams['figure.figsize'] = (9.0, 6.0)
from sklearn.svm import SVC
from skimage.util.shape import view_as_windows
import pickle
from skimage import measure
from skimage.filters import gaussian_filter

size = 50
var = 0.99

f = open('class_stack_'+str(size)+'x'+str(size)+'_var0'+str(int(var*100e0))+'.dat','r')
[class_stack] = pickle.load(f)
f.close()

print np.shape(class_stack)
```

```python
from skimage.filters.rank import entropy
from skimage.morphology import disk

frames = np.arange(0,10)*100

#for f in frames:
for f in range(np.shape(class_stack)[0]):
    smoothed = gaussian_filter(class_stack[f,:,:],sigma=10)
    
    contours = measure.find_contours(smoothed, 0.5)
    
    plt.axis('equal')
    plt.xlim([0,np.shape(image_stack[f,:,:])[0]])
    plt.ylim([0,np.shape(image_stack[f,:,:])[1]])

    print len(contours)
    
    if len(contours) != 0:
    
        # calculate the area of the contours
        area = np.zeros(len(contours))

        for n, contour in enumerate(contours):
            x = contour[:,0]
            y = contour[:,1]
            # copy the first point to the end of the contour list to close the loop
            x = np.append(x,x[0])
            y = np.append(y,y[0])
            # use Green's theorem to calculate the area of the contour
            area[n] = np.abs(np.sum(x[:-1]*y[1:]-y[:-1]*x[1:]))/2e0

        # find contour with the largest area
        cont = contours[np.where(area == np.max(area))[0]]

        plt.plot(size/2e0+cont[:, 1], size/2e0+cont[:, 0],color='r', linewidth=2) 
        #for n, contour in enumerate(contours):
        #    plt.plot(size/2e0+contour[:, 1], size/2e0+contour[:, 0],color='r', linewidth=1) 

    plt.imshow(class_stack[f,:,:],cmap='Greys_r')
    plt.savefig('animation_border/contours_tiles_frame'+str(f).zfill(4)+'_size'+str(size)+'.png',dpi=150)
    plt.close()

    
```

```python
from skimage.filters.rank import entropy
from skimage.morphology import disk

frames = np.arange(0,10)*100

for f in frames:
    
    enthropy_image = entropy((image_stack[f,:,:] - np.max(image_stack[f,:,:]))/(np.min(image_stack[f,:,:]) - np.max(image_stack[f,:,:])), disk(5))
    
    
    level = np.percentile(enthropy_image,75)
    
    contours = measure.find_contours(enthropy_image, level)

    # calculate the area of the contours
    area = np.zeros(len(contours))

    for n, contour in enumerate(contours):
        x = contour[:,0]
        y = contour[:,1]
        # copy the first point to the end of the contour list to close the loop
        x = np.append(x,x[0])
        y = np.append(y,y[0])
        # use Green's theorem to calculate the area of the contour
        area[n] = np.abs(np.sum(x[:-1]*y[1:]-y[:-1]*x[1:]))/2e0

    # find contour with the largest area
    cont = contours[np.where(area == np.max(area))[0]]

    plt.axis('equal')
    plt.xlim([0,np.shape(image_stack[f,:,:])[0]])
    plt.ylim([0,np.shape(image_stack[f,:,:])[1]])
    plt.plot(cont[:, 1], cont[:, 0],color='r', linewidth=3) 
    for n, contour in enumerate(contours):
        plt.plot(contour[:, 1], contour[:, 0],color='r', linewidth=1) 

    
    
    
    plt.imshow(image_stack[f,:,:],cmap='Greys_r')
    plt.savefig('imgs/enthropy_frame'+str(f)+'_size'+str(size)+'.png',dpi=150)
    plt.show()

    
```
